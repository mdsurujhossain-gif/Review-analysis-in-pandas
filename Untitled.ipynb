## ploting average rating / rating count/min rating/max rating by day 

```python
import pandas
from datetime import datetime
import matplotlib.pyplot as plt
from pytz import utc
data = pandas.read_csv('reviews.csv',parse_dates=['Timestamp'])
data.head()
```

```python
data['Day'] = data['Timestamp'].dt.date
##day_count = data.groupby(['Day']).count()
##day_count
day_average = data.groupby(['Day']).mean()
day_average
```

```python
plt.figure(figsize=(25,3))
plt.plot(day_average.index,day_average['Rating'])
day_average['Rating'].max()
## day_average['Rating'].min()
```

## average rating by week

```python
data['Week'] = data['Timestamp'].dt.strftime('%Y-%U')
week_average = data.groupby(['Week']).mean()
##Week_average
plt.figure(figsize=(25,5))
plt.plot(week_average.index,week_average['Rating'])
```

## downsampling and average rating by month

```python
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')
monthly_average = data.groupby(['Month']).mean()
##mohtly average
plt.figure(figsize=(25,5))
plt.plot(monthly_average.index,monthly_average['Rating'])
```

## average rating by month by course

```python
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')
monthly_average_course = data.groupby(['Month','Course Name'])['Rating'].mean().unstack()
monthly_average_course.plot(figsize=(25,8))
##monthly_average_course[:20]
```
